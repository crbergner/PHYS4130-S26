```python
'''
Filename: finaldraft.ipynb
Written by: Cricket Bergner
Date: 04/27/26
'''

# import libraries
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import solve_ivp as intg    # for integrate
import random as ra

# functions

def acceleration(x, y, vx, vy, magnets):
    ax = -restoring * x - damping * vx
    ay = -restoring * y - damping * vy

    for mag in magnets:
        dx = mag[0] - x
        dy = mag[1] - y
        r = np.sqrt(dx**2 + dy**2 + vertical_offset**2)
        ax += magnetic_strength * dx / r**3
        ay += magnetic_strength * dy / r**3

    return ax, ay

def equations(t, state, magnets):
    x, y, vx, vy = state
    ax, ay = acceleration(x, y, vx, vy, magnets)
    return [vx, vy, ax, ay]

def rk45(initial_state, magnets):
    solution = intg(
        fun = lambda t, y: equations(t, y, magnets),
        t_span=t_span,
        y0=initial_state,
        t_eval=t_eval,
        method='RK45')

    return solution

# initial variables

damping = 0.2
restoring = 0.6
vertical_offset = 0.3
magnetic_strength = 1.2

t_span = (0,20)
t_eval = np.linspace(0, 20, 2048)

initial_conditions, solutions = [], []

# main body (3 magnets)

# place three magnets on the ends of an equilateral triangle
magnets3 = [np.array([1, 0]),
          np.array([-0.5, np.sqrt(3)/2]),
          np.array([-0.5, -np.sqrt(3)/2])]

for _ in range(3): # randomly generate initial conditions
    x0 = ra.uniform(-0.5, 0.5)
    y0 = ra.uniform(-0.5, 0.5)
    vx0 = ra.uniform(-0.1, 0.1)
    vy0 = ra.uniform(-0.1, 0.1)
    initial_conditions.append([x0, y0, vx0, vy0])

for i in initial_conditions:
    ans = rk45(i, magnets3)
    solutions.append(ans)

# plotting
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, j in enumerate(solutions):
    ax = axes[i]
    x = j.y[0]
    y = j.y[1]

    ax.plot(x, y, lw=0.7)

    # plot magnets
    for m in magnets3:
      ax.scatter(m[0], m[1], color='red', s=50, zorder=3)

    ic = initial_conditions[i]
    ax.set_title(
        f"x0={ic[0]:.2f}, y0={ic[1]:.2f}\n"
        f"vx0={ic[2]:.2f}, vy0={ic[3]:.2f}"
    )
    ax.set_aspect('equal')

    ax.grid(alpha=0.3)
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)

plt.suptitle("Magnetic Pendulum Trajectories (3 Random Initial Conditions)")
#plt.tight_layout() <- seems to be causing a formatting issue
plt.show()



# extension one (4 magnets)

# place four magnets on the ends of a square
# magnets4 = [np.array([1, 0]),
#             np.array([1, 1]),
#             np.array([0, 1]),
#             np.array([0, 0])]


# extension two (5 magnets)






#################################################################
# END FINAL PROJECT
#################################################################
```
